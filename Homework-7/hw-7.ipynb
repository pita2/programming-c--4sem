```python
from scipy.spatial.distance import cdist, pdist
from scipy.cluster import hierarchy
import warnings
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=10**3,
                       centers=3,
                       cluster_std=1)
X = X[:, ::-1]

# Строим точки
plt.figure(1)
colors = [ 'skyblue', 'red', 'y']

for k, col in enumerate(colors):
    cluster_data = y == k
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1],
                c=col, marker='.', s=1)
```

```python
distance_mat = pdist(X, metric='sqeuclidean')
Z = hierarchy.linkage(distance_mat, 'complete')
plt.figure(figsize=(10,6))
dn = hierarchy.dendrogram(Z)
```

```python
from sklearn.cluster import kmeans_plusplus
plt.figure(1)
colors = ['pink', 'skyblue', 'red', 'navy', 'y', 'black', 'magenta']

for k, col in enumerate(colors):
    cluster_data = y == k
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1],
                c=col, marker='.', s=0.5)

# kmeans work
centers_init, indices = kmeans_plusplus(X, n_clusters=3,
                                        random_state=0)

plt.scatter(centers_init[:, 0], centers_init[:, 1], c='b', s=50)
plt.xticks([])
plt.yticks([])
plt.show()
```
